```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
import matplotlib.pyplot as plt
from io import StringIO
# URL of the Pokémon type chart
url = 'https://pokemondb.net/type'

# Fetch the content of the webpage
response = requests.get(url)
webpage_content = response.content

# Parse the HTML content
soup = BeautifulSoup(webpage_content, 'html.parser')


# For example, if the chart is in a table with a specific ID or class, you would use soup.find() or soup.find_all()
type_chart = soup.find('table', class_='type-table')


data = []
for row in type_chart.find_all('tr')[1:]:  
    columns = row.find_all('td')
    type_name = columns[0].text.strip()  
    effectiveness = [col.text.strip() for col in columns[1:]]  
    data.append({
        'Type': type_name,
        'Effectiveness': effectiveness  # This is a simplification; adjust as needed
    })  # Populate with actual data

```

```python
types = [ 'Normal', 'Fire', 'Water', 'Electric', 'Grass', 'Ice', 'Fighting', 'Poison', 'Ground', 'Flying', 'Psychic', 'Bug', 'Rock', 'Ghost', 'Dragon', 'Dark', 'Steel', 'Fairy']

# Populate the DataFrame 
eff =  [[entry['Type']] + entry['Effectiveness'] for entry in data]
df_effectiveness = pd.DataFrame(eff)
df_effectivenesscolumns = types
df_effectivenessindex = types

# Replace empty strings with NaN for clarity
df_effectiveness = df_effectiveness.replace('', pd.NA)

df_effectiveness.to_csv('pokemon_effectiveness.csv', index=False)
```

```python

# URL of the page
url = 'https://bulbapedia.bulbagarden.net/wiki/List_of_Pokémon_by_effort_value_yield_(Generation_IV)'

# Fetch the content of the webpage
response = requests.get(url)
webpage_content = response.content

# Parse the HTML content
soup = BeautifulSoup(webpage_content, 'html.parser')

# Find the table with the required class
ev_table = soup.find('table', class_='sortable roundy')

# Initialize a list to store your data
pokemon_ev_data = []

# Iterate over the rows in the table, skipping the header row
for row in ev_table.find_all('tr')[1:]:
    cols = row.find_all('td')
    # Extract the data from the columns
    dex_number = cols[0].text.strip()
    name = cols[2].text.strip()
    exp = cols[3].text.strip()
    hp_ev = cols[4].text.strip()
    attack_ev = cols[5].text.strip()
    defense_ev = cols[6].text.strip()
    sp_attack_ev = cols[7].text.strip()
    sp_defense_ev = cols[8].text.strip()
    speed_ev = cols[9].text.strip()

    pokemon_ev_data.append({
        'Dex Number': dex_number,
        'Name': name,
        'EXP': exp,
        'HP EV': hp_ev,
        'Attack EV': attack_ev,
        'Defense EV': defense_ev,
        'Sp. Attack EV': sp_attack_ev,
        'Sp. Defense EV': sp_defense_ev,
        'Speed EV': speed_ev
    })

# Create a DataFrame from the list of dictionaries
df_pokemon_ev = pd.DataFrame(pokemon_ev_data)

# Display the first few rows of the DataFrame to ensure it's loaded correctly
print(df_pokemon_ev.head())
df_pokemon_ev.to_csv('Pokemon_EV_EXP.csv',index=False)
```
